```python
#above is all exploring the dataset, now we should consider how we prepare the dataset for our prediction modeling

#how do we split data
#depends on dependency structure and homogeneity
#some ways to split:
#1) time series data -> best way to split is in 
#order of time (only use observations that have occurred in past to predict future values)
#failure to account for time-series could result in a biased estimate of 
#generalization error caused by look-ahead (leakage) bias
#in this example time is not an issue


#2) cross-sectional and heterogeneous data -> when subgroups of data are very similar (homogeneous)
#as measured by some similarity measure (ie. correlations)
#best way to. split is using stratified sampling technique (as opposed to random sampling)
#stratified sampling is when the population is divided into similar (homogeneous) subgroups called strata
#and a representative number of observations are taken (sampled) from each stratum to gurantee 
#the test is representative of overall population
#basically you split groups into homogenous groups (like groups with similar qualities) 
#and then you take a little from each group to represent the whole population 
#we might consider sampling based off median_income groups
```

# What's an example of look-ahead bias and sampling bias?
# look-ahead: making a decision based on info you would only get in the future
# ie) selling a stock now because your data tells you that it will drop in the future



# sampling bias is just not correctly taking a representative sample
# ie) we could have a population of 25 freshman, 25 soph, 25 juniors, 25 seniors
# our random sample might take 15 freshman, 5 soph, 2 jun, 1 sen
# not representative of the population

```python
# If you decide you want to ensure that the test set is representative of 
#the various categories of incomes in the whole data set, you may want to use stratified random sampling.

# Since the median income is a continuous numerical value, you first need to create an 
#discrete income category variable.


# It is important to have a sufficient number of observations in your data 
#set for each stratum, otherwise the estimate of a stratum's importance may be biased.
# This implies you shouldn't have too many strata, and each stratum should be large enough.
```

```python
# One way to evaluate the Decision Tree model would be to use the 
# train_test_split() method to split the training set into a smaller training and validation set. 
# Then train your models against the smaller training set and evaluate them against the validation set.

# A great alternative is Scikit-Learn's K-fold cross-validation functionality.

# We will split the training set into 10 distinct subsets called folds.

# Subsequently, the model gets estimated and evaluated 10 times, picking a different fold for 
# evaluation every time while training on the other 9 folds.

# The result is an array containing the 10 evaluation scores
```

```python
# Cross-validation can often be a bottleneck to you code
# Sklearn's implementation of cross validation scores expects a "utility/gain function" 
# rather than a "loss/cost function"

from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, X_train_cleaned, Y_train, 
                                scoring="neg_mean_squared_error",
                                cv=10)

tree_rmse_in_sample_estimates = np.sqrt(-scores)

tree_rmse_in_sample_estimates
```

```python
# To calculate a 95% confidence interval for the point estimate of the generalization out-of-sample error, 
# we'll use Scipy and more-or-less do the computation manually.

from scipy import stats

confidence = 0.95
squared_errors_oos = (Y_test - Y_hat_test)**2

confidence_interval = \
np.sqrt(stats.t.interval(alpha=confidence,                     # confidence level
                         df=len(squared_errors_oos) -1,        # degrees of freedom
                         loc=squared_errors_oos.mean(),        # location of the t-distribution
                         scale=stats.sem(squared_errors_oos))) # scale of the t-distribution set to be the 
                                                               # std error of the mean of the out-of-sample squared errors
```

```python
print(f'The point estimate for the average test set (OOS) error is {np.round(np.sqrt(squared_errors_oos.mean()),0)}\
 and the confidence interval is {np.round(confidence_interval,0)}.')
```
